## Data Cleaning

```python
#remove the column (Notes) because it is empty
df.drop("Notes" ,axis=1, inplace=True)
df.drop("Status" ,axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
#Fill all NAN values with median of it's column
df[['BasePay', 'OvertimePay', 'OtherPay']] = df[['BasePay', 'OvertimePay', 'OtherPay']].fillna( df[['BasePay', 'OvertimePay', 'OtherPay']].median())
df['Benefits']=df['Benefits'].fillna(0)
df.isnull().sum()
```

# Basic Data Visualization

```python
#histogram of salaries
%matplotlib inline
import matplotlib.pyplot as plt
df['TotalPay'].hist(bins=50, figsize=(20,15))
plt.show()
```

```python
import matplotlib.pyplot as plt
# Get the number of employees in each department
department_counts =df['JobTitle'].value_counts()

# Create a pie chart
plt.pie(department_counts, labels=department_counts.index)
plt.title("Proportion of Employees in Different Departments")
plt.show()
```

# Grouped Analysis

```python
#create a group and find avg of salary
avg_salary_by_jobtitle=df.groupby('JobTitle')['TotalPayBenefits'].mean()
print(avg_salary_by_jobtitle)
```

# Simple Correlation Analysis

```python
import matplotlib.pyplot as plt
# Create a scatter plot to visualize the relationship between salary after benefits through different years
plt.scatter(df['Year'], df['TotalPayBenefits'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Salary vs Years of Experience')
plt.show()
```
